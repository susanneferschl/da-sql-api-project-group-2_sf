```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'weather_flights_x_sf' DataFrame is already defined and contains relevant data

# Convert 'flight_date' column to datetime format
weather_flights_x_sf['flight_date'] = pd.to_datetime(weather_flights_x_sf['flight_date'], format='%y:%m:%d')

# Sort the DataFrame by 'flight_date'
weather_flights_x_sf_sorted = weather_flights_x_sf.sort_values(by='flight_date')

# Convert 'cancelled' column to numerical type (assuming it's boolean)
weather_flights_x_sf_sorted['cancelled'] = weather_flights_x_sf_sorted['cancelled'].astype(int)

# Group by flight date and origin, and calculate mean snowfall and cancellation rate
grouped_data = weather_flights_x_sf_sorted.groupby(['flight_date', 'origin']).agg({'snow': 'mean', 'cancelled': 'mean'}).reset_index()

# Calculate overall cancellation rate
overall_cancellation_rate = (weather_flights_x_sf_sorted['cancelled'] == 1).mean() * 100

# Create subplots for each origin
unique_origins = weather_flights_x_sf_sorted['origin'].unique()
num_origins = len(unique_origins)

fig, axes = plt.subplots(num_origins, 2, figsize=(20, 6*num_origins))
plt.suptitle('Weather conditions per day by flight origin', fontsize=20)

for i, origin in enumerate(unique_origins):
    # Filter data for current origin
    data_origin = grouped_data[grouped_data['origin'] == origin]
    
    # Barplot for snow per day
    sns.barplot(x='flight_date', y='snow', data=data_origin, ax=axes[i, 0])
    axes[i, 0].tick_params(axis='x', labelrotation=90)
    axes[i, 0].set_ylabel("Snow in mm", fontsize=10)
    axes[i, 0].set_title(f"Mean Snowfall per day - Origin: {origin}", fontsize=15)

    # Plot cancellation rate per origin
    cancellation_rate_origin = data_origin['cancelled'].mean() * 100
    axes[i, 1].bar(['Cancellation Rate'], [cancellation_rate_origin], color='red')
    axes[i, 1].set_ylabel("Cancellation Rate (%)", fontsize=10)
    axes[i, 1].set_title(f"Cancellation Rate - Origin: {origin}", fontsize=15)

plt.tight_layout()
plt.show()
```
