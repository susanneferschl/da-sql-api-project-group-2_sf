```python
schema = 'hh_analytics_24_1'

#sql_query to retrieve the data from DBeaver and saving it as a df
sql_query = f'SELECT * FROM {schema}.weather_stacked_groupx'
weather_stacked_sf = get_dataframe(sql_query)

weather_stacked_sf
```
