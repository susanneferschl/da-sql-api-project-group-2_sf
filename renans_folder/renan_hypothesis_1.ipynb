```python
# The columns in the database have different naming as in the source csv files. Lets get the names from the database

schema = 'hh_analytics_24_1' # UPDATE 'TABLE_SCHEMA' based on schema used in class 
engine = get_engine() # assign engine to be able to query against the database

table_name_sql = f'''SELECT COLUMN_NAME 
                    FROM INFORMATION_SCHEMA.COLUMNS 
                    WHERE TABLE_NAME = 'flights'
                    AND TABLE_SCHEMA ='{schema}'
                    ORDER BY ordinal_position'''
c_names = engine.execute(table_name_sql).fetchall()
c_names
```

```python
# we can clean up the results into a clean list
new_column_names=[]
for name in c_names:
    new_column_names.append(name[0])
new_column_names        
```

```python
def clean_airline_df(df):
    '''
    Transforms a df made from BTS csv file into a df that is ready to be uploaded to SQL
    Set rows=0 for no filtering
    '''

    # Build dataframe including only the columns you want to keep
    df_airline = df.loc[:,columns_to_keep]
     
    # Clean data types and NULLs
    df_airline['FlightDate']= pd.to_datetime(df_airline['FlightDate'], yearfirst=True)
    df_airline['CRSArrTime']= pd.to_numeric(df_airline['CRSArrTime'], downcast='integer', errors='coerce')
    df_airline['Cancelled']= pd.to_numeric(df_airline['Cancelled'], downcast='integer')
    df_airline['Diverted']= pd.to_numeric(df_airline['Diverted'], downcast='integer')
    df_airline['ActualElapsedTime']= pd.to_numeric(df_airline['ActualElapsedTime'], downcast='integer', errors='coerce')
    
    # Rename columns
    df_airline.columns = new_column_names
    
    return df_airline
```

```python
# Call function and check resulting dataframe
df_clean = clean_airline_df(df)
df_clean.head()
```

```python
# Specify the airports you are interested in and put them as a list in the function.
def select_airport(df, airports):
    ''' Helper function for filtering the airline dataframe for a subset of airports'''
    df_out = df.loc[(df.origin.isin(airports)) | (df.dest.isin(airports))]
    return df_out
```

```python
# Execute function, filtering for New York area airports
airports=['DCA', 'IAD', 'BWI', 'PHL', 'EWR', 'JFK', 'LGA']
if len(airports) > 0:
    df_selected_airports = select_airport(df_clean, airports)
else:
    df_selected_airports = df_clean
    
df_selected_airports.tail()
```

```python
#Deep dive in cancellations: show count of cancelled flights in origin airports

import pandas as pd

# Assuming you have already loaded your data into a DataFrame called df_flights_2010_2

# Define the list of selected airports
selected_airports = ['DCA', 'IAD', 'BWI', 'PHL', 'EWR', 'JFK', 'LGA']

# Extract the year from the flight_date column
df_selected_airports['year'] = pd.to_datetime(df_selected_airports['flight_date']).dt.year

# Filter the DataFrame to include only the flights from the year 2010
df_selected_airports_2010 = df_selected_airports[df_selected_airports['year'] == 2010]

# Filter the DataFrame to include only the selected airports
df_selected_airports_2010 = df_selected_airports_2010[df_selected_airports_2010['origin'].isin(selected_airports)]

# Group by 'origin' and count the number of flights for each airport
departure_count = df_selected_airports_2010.groupby('origin')['flight_number'].count().reset_index()
departure_count.columns = ['origin', 'departure_count']

# Calculate count of cancelled flights
cancelled_count = df_selected_airports_2010.groupby('origin')['cancelled'].sum().reset_index()
cancelled_count.columns = ['origin', 'cancelled_count']

# Merge departure_count and cancelled_count
departure_cancelled_count = departure_count.merge(cancelled_count, on='origin', how='left')

# Calculate airport rank based on departure count
departure_cancelled_count['airport_rank'] = departure_cancelled_count['departure_count'].rank(ascending=False).astype(int)

# Sort the result by departure count in descending order
departure_cancelled_count = departure_cancelled_count.sort_values('departure_count', ascending=False)

# Reset index for display
departure_cancelled_count = departure_cancelled_count.reset_index(drop=True)

print(departure_cancelled_count)
```

ii. Calculate cancellation rates

```python
import pandas as pd

# Assuming you have already loaded your data into a DataFrame called df_flights_2010_2

# Define the list of selected airports
selected_airports = ['DCA', 'IAD', 'BWI', 'PHL', 'EWR', 'JFK', 'LGA']

# Extract the year from the flight_date column
df_selected_airports['year'] = pd.to_datetime(df_selected_airports['flight_date']).dt.year

# Filter the DataFrame to include only the flights from the year 2010
df_selected_airports_2010 = df_selected_airports[df_selected_airports['year'] == 2010]

# Filter the DataFrame to include only the selected airports
df_selected_airports_2010 = df_selected_airports_2010[df_selected_airports_2010['origin'].isin(selected_airports)]

# Group by 'origin' and count the number of flights for each airport
departure_count = df_selected_airports_2010.groupby('origin')['flight_number'].count().reset_index()
departure_count.columns = ['origin', 'departure_count']

# Calculate count of cancelled flights
cancelled_count = df_selected_airports_2010.groupby('origin')['cancelled'].sum().reset_index()
cancelled_count.columns = ['origin', 'cancelled_count']

# Merge departure_count and cancelled_count
departure_cancelled_count = departure_count.merge(cancelled_count, on='origin', how='left')

# Calculate cancellation rate (%)
departure_cancelled_count['cancellation_rate(%)'] = (departure_cancelled_count['cancelled_count'] / departure_cancelled_count['departure_count']) * 100

# Calculate airport rank based on departure count
departure_cancelled_count['airport_rank'] = departure_cancelled_count['departure_count'].rank(ascending=False).astype(int)

# Sort the result by departure count in descending order
departure_cancelled_count = departure_cancelled_count.sort_values('departure_count', ascending=False)

# Reset index for display
departure_cancelled_count = departure_cancelled_count.reset_index(drop=True)

print(departure_cancelled_count)
```

## Part 2 Visualisation 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already loaded your data into a DataFrame called df_selected_airports
# Extract year from flight_date
df_selected_airports['year'] = df_selected_airports['flight_date'].dt.year

# Calculate flight count and cancellation rate
result_df = df_selected_airports.groupby('year').agg(
    flight_count=('flight_date', 'size'),
    cancellation_rate=('cancelled', lambda x: round((x.sum() / len(x)) * 100, 2)),
    flight_cancelled_count=('cancelled', 'sum')  # Summing up cancelled flights
).reset_index()

# Sort the result by year
result_df = result_df.sort_values('year')

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 6))

# Stacking flight count and cancelled flight count
ax1.bar(result_df['year'], result_df['flight_count'], color='skyblue', label='Flight Count')
ax1.bar(result_df['year'], result_df['flight_cancelled_count'], color='salmon', label='Cancelled Flight Count', bottom=result_df['flight_count'])

# Adding data labels for flight count and cancelled flight count
for i, v in enumerate(result_df['flight_count']):
    ax1.text(result_df['year'][i], v + result_df['flight_cancelled_count'][i] / 2, str(int(v)), color='black', ha='center', va='bottom')
    ax1.text(result_df['year'][i], v / 2, str(result_df['flight_cancelled_count'][i]), color='black', ha='center', va='bottom')

# Setting title and labels
plt.title('Count of Flights in February Across 2010-2017')
ax1.set_xlabel('Year')
ax1.set_ylabel('Flight Count')
ax1.tick_params(axis='y')

# Removing grid lines
ax1.grid(False)

# Adding secondary y-axis for cancellation rate
ax2 = ax1.twinx()
ax2.plot(result_df['year'], result_df['cancellation_rate'], color='orange', label='Cancellation Rate (%)', marker='o')
ax2.set_ylabel('Cancellation Rate (%)')

# Adding data labels for cancellation rate
for i, v in enumerate(result_df['cancellation_rate']):
    ax2.text(result_df['year'][i], v, f'{v:.2f}%', color='black', ha='center', va='bottom')

# Adding legend
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper center', bbox_to_anchor=(0.88, 1.0), fontsize='small')

# Show plot
plt.show()
```
